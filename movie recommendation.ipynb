```python
import pandas as pd
import numpy as np


data = pd.read_csv('movies.csv')
data.head(10)


data.shape
data.info()


data.columns

movies=data[[   'id', 'genres' , 'keywords', 'original_language','overview',
       'original_title', 'popularity', 'production_companies', 'release_date', 'runtime',
       'title'    ]]


movies

movies.info()
movies.isnull().sum()
movies.dropna(axis=0, how= 'any')


df = pd.read_csv('credits.csv')

df.shape

df.info()

df.isnull().sum()

movies = movies.merge(df, on = 'title')
movies


movies.iloc[0:4801].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L


movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['production_companies'] = movies['production_companies'].apply(convert)


movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
        return L


movies['cast'] = movies['cast'].apply(convert3)
movies
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
        return L


movies['crew'] = movies['crew'].apply(fetch_director)

movies
```

```python
movies['overview'][0]
```

```python
def safe_split(x):
    if isinstance(x, str):
        return x.split()
    else:
        return [] 

movies['overview'] = movies['overview'].apply(safe_split)

movies
```

```python


movies=movies.dropna()

```

```python
movies.isnull().sum()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])

movies

movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" , ","") for i in x])
movies
```

```python
movies['overview'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])


movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies
```

```python
new_df = movies[['id','title','tags']]
new_df

```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x)if isinstance(x, list) else str(x)) 
new_df
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)

cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    dictance=similarity[movie_index]
    movies_list = sorted(list(enumerate(dictance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index
    if movie_index.empty:
        print(f"Movie '{movie}' not found in the dataset.")
        return  # Exit the function if the movie is not found

    movie_index = movie_index[0]  # Get the first (and only) index
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('justice league')
```
